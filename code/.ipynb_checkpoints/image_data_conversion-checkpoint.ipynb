```python
import pandas as pd
from matplotlib import pyplot as plt


def import_pos(t_start,t_end):
    data=pd.read_csv("~/year_2/optimal_transport/data/particle_positions.csv")

    data.rename(columns = {'frame' : 't', 'x [nm]' : 'x','y [nm]' : 'y','intensity [photon]' : 'intensity'}, inplace=True)

    data=data.loc[(data.t>=t_start & data.t<=t_end)]

    return data
```
