```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("~/year_2/optimal_transport/data/particle_positions.csv")
data
```

```python
n_frames=100
data=data.loc[data.frame<=n_frames]
```

```python
relevant_frames=['frame','x [nm]','y [nm]']
data=data[relevant_frames]
```

```python
plt.scatter(data['x [nm]'].to_numpy(),data['y [nm]'].to_numpy())
```

```python
arr=data.to_numpy()
```

```python
import btrack
import napari
from matplotlib import pyplot as plt
```

```python
objects = btrack.dataio.objects_from_array(arr, ['t','x', 'y'])
objects
```

```python
config = '../cell_config.json'
```

```python
with btrack.BayesianTracker() as tracker:
    # load the configuration file
    tracker.configure_from_file(config)
    #tracker.max_search_radius = 80*10
    # add the objects to the tracker
    tracker.append(objects)
    tracker.volume=((0, 80*368), (0,80*358), (-1e6, 1e6))
    # tracl objects
    tracker.track_interactive(step_size=1)
    # run the global optimizer
    tracker.optimize()
    # tracks as a python list
    tracks = tracker.tracks
    # get tracks formated for napari
    data, properties, graph = tracker.to_napari(ndim=2)
```

```python
for trck in tracks:
    plt.plot(trck.x,trck.y)
```
